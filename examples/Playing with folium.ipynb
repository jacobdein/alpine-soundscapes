```python
from database.models import Site
import folium
```

```python
sites = Site.objects.all()
```

```python
site15 = Site.objects.get(name = 'Hofgarten')
```

```python
map_sites = folium.Map(location=[site15.lat, site15.lon], zoom_start=12, tiles='Stamen Terrain')
```

```python
for site in sites:
    folium.Marker([site.lat, site.lon], popup=site.name).add_to(map_sites)
```
